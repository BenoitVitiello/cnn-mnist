# 1 Exploration

This notebook explores the dataset before CNN training.

1. Data structure
2. Data visualization
3. Class distribution
4. Pixel stats

## 0 - Imports

```python
import numpy as numpy
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.datasets import mnist
```

## 1 - Data loading

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"X_train : {X_train.shape}")
print(f"y_train : {y_train.shape}")
print(f"X_test  : {X_test.shape}")
print(f"y_test  : {y_test.shape}")
```

## 2 - Data visualization

```python
fig, axes = plt.subplots(2, 5, figsize = (12, 5))
fig.suptitle("Exemples d'images MNIST", fontsize = 14)

for i, ax in enumerate(axes.flat):
    ax.imshow(X_train[i], cmap = "gray")
    ax.set_title(f"Label : {y_train[i]}")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

## 3 - Class distribution

```python
fig, axes = plt.subplots(1, 2, figsize = (14, 5))

for ax, labels, title in zip(axes, [y_train, y_test], ["Train", "Test"]):
    sns.countplot(x = labels, hue = labels, ax = ax, palette = "Blues_d", legend = False)
    ax.set_title(f"Class distribution — {title}")
    ax.set_xlabel("Label")
    ax.set_ylabel("Occurences")

plt.tight_layout()
plt.show()
```

## 4 - Pixel stats

```python
print(f"Min pixel value : {X_train.min()}")
print(f"Max pixel value : {X_train.max()}")
print(f"Mean : {X_train.mean():.2f}")
print(f"Standard deviation : {X_train.std():.2f}")
```
